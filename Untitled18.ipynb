```python
import numpy as np
import pandas as pd 
```

```python
file=pd.read_csv('/content/prototype_2.csv')
file.head()
```

```python
file.shape
file.isnull().sum()
```

```python
newfile=file.dropna()
newfile.shape
```

```python
from sklearn import preprocessing
```

```python
lat_long=newfile[['latitude','longitude']]
```

```python
scaler = preprocessing.StandardScaler()
standardized_x = scaler.fit_transform(lat_long)
print(standardized_x)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data=lat_long ,x="latitude",y="longitude")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
kmeans = KMeans(3)
y_predicted=kmeans.fit_predict(lat_long)
```
